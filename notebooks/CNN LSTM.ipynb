**Reference:** https://github.com/vinayakumarr/Network-Intrusion-Detection

**Dataset:** http://kdd.ics.uci.edu/databases/kddcup99/kddcup99.html

```python
import numpy as np
import pandas as pd

import tensorflow as tf

from utils import preprocess
```

```python
DATA = '../data/'
MODELS = '../models/'
TEST = 0.3
```

```python
traindata = pd.read_csv(DATA+'kddcup.data.corrected.csv', header=None)
traindata
```

```python
X_train, X_test, y_train, y_test = preprocess.preprocess(traindata)
```

```python
X_train
```

```python
y_train.columns
```

```python
X_train, X_test = np.array(X_train), np.array(X_test)
y_train, y_test = np.array(y_train), np.array(y_test)
```

```python
# reshape input to be [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv1D(8, 5, padding="same", activation="relu", input_shape=X_train.shape[1:]))
model.add(tf.keras.layers.MaxPooling1D(2))
model.add(tf.keras.layers.LSTM(8))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(y_train.shape[1], activation="softmax"))
```

```python
# define optimizer and objective, compile model
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
model.summary()
```

```python
# train
model.fit(X_train, y_train, epochs=3, validation_split=0.4, shuffle=False, batch_size=128)
```

```python
model_path = MODELS+"cnn_lstm_model.hdf5"
```

```python
model.save(model_path)
```

```python
model = tf.keras.models.load_model(model_path)
```

```python
model.evaluate(X_test, y_test, batch_size=128)
```
